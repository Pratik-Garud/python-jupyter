Classify employees based on performance ratings and store classifications in a new dictionary.
If >90 = Excellent
If >75 = Good
Else Need Improvement
employee_ratings = {"John": 85, "Alice": 90, "Bob": 70, "Daisy": 95}

```python
employee_ratings = {"John": 85, "Alice": 90, "Bob": 70, "Daisy": 95}

Excellent = {}
Good = {}
Need_Improvment= {}

for key,value in employee_ratings.items():
    if value > 90:
        Excellent[key] = value
    elif value > 75:
        Good[key] = value
    else:
        Need_Improvment[key] = value
```

Given a list of daily complaint counts, calculate the weekly average and print days with complaints above this average. 
daily_complaints = [15, 20, 10, 25, 30, 5, 12]

```python
daily_complaints = [15, 20, 10, 25, 30, 5, 12]
avg_count = 0
sum_count = 0
for i in daily_complaints:
     sum_count += i
avg_count = sum_count / len(daily_complaints)
print("weekly avg complaints: ",round(avg_count,2))
for i,val in enumerate(daily_complaints,start =1):
    if val > avg_count:
        print(f"day {i} : {val}")
```
