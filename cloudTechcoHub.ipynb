```python
try:
    x= 10/0
    print(x)
except ZeroDivisionError:
    print("dont divid by zero")
except IndentationError:
    print("An exception ocure")
```

```python
def divi(a,b):
    try:
        x= a/b
        print(x)
    except ZeroDivisionError:
         print("dont divid by zero")
    except IndentationError:
        print("An exception ocure")
    except TypeError:
        print("insert only numerical values")
    except:
        print("somthing gone wrong")
    finally:
        print("the try and except is finished")

divi(5,0)
```

```python
data = {
    "month" : ["jan","feb" ," mar"],
    "product1" : [123,165,198],
    "product2" : [202,120,151]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=df,x="month", y="product1", label = "product1",color="blue")
sns.lineplot(data=df,x="month", y="product2", label = "product2",color="black")
plt.title("sales")
plt.xlabel("month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df,x="month", y="product1", label = "product1")
sns.barplot(data=df,x="month", y="product2", label = "product2")
plt.title("sales")
plt.xlabel("month")
plt.ylabel("date")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
data= {
    "month" : ["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"],
    "Adv_spend" : [20,25,30,35,40,45,50,55,60,65,70,75],
    "sales" : [200,250,270,300,330,350,370,380,400,420,440,460]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
print(df)
```

```python
x = df[["Adv_spend"]]
y = df[["sales"]]
x_train,x_test,y_train,y_test =   train_test_split(x,y,test_size = 0.2,random_state = 42)
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
new_adv_spend = np.array([[50]])
pred_sales = model.predict(new_adv_spend)
print("pre sales:" , pred_sales)
```

```python
new_adv_spend = np.array([[70]])
pred_sales = model.predict(new_adv_spend)
print("pre sales:" , pred_sales)
```
