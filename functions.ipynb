```python
def recursion_sum(list_of_numbers):
    print(list_of_numbers)
    if len(list_of_numbers) == 1:
        return list_of_numbers[0]
    else:
        return list_of_numbers[0] + recursion_sum(list_of_numbers[1:])
```

```python
recursion_sum([1,2,3,4,5])
```
