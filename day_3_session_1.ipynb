```python
data = """
A, B, C; D, E, F
G, H, I
J, K, L; M, N, O
"""
```

```python
data1 = re.split('[\n;+]',data)
data1
```

```python
data1 = data1.strip()
```

```python
data1 = re.split('[\n;+]',data.strip())
```

```python
data1
```

```python
data2 = re.split("[';\n+]",data.strip())
data2
```

```python
table = [re.split(r'\s*,\s*', row.strip()) for row in data1]
print(table)
```

Explanation:

We’re looping through each row (for row in rows).

Inside each row, we use re.split(r'\s*,\s*', row.strip()).

\s* means “zero or more spaces”.

So it splits on commas, even if they have spaces before or after.

.strip() again removes leftover spaces.
