```python
def marks(t_marks,p_marks,b_marks = 5):
    t_marks = t_marks + p_marks + b_marks
    if t_marks > 100:
        t_marks = 100
    print(f"total score is {t_marks}")
```

```python
marks(50,40)
```

```python
marks(40,60,10)
```

```python
marks(40,70)
```

```python
def check(n):
    if n>0:
        print(f" {n} is positive")
    elif n<0:
        print(f" {n} is negative")
    else:
        print(f" {n} is zero")

    if n%2 == 0:
        print(f" {n} is even")
    else:
        print(f" {n} is odd")

    if n%3 == 0:
        print(f" {n} is divisible by 3")
    else:
        print(f" {n} is not divisible by 3")
        
    if n%5 == 0:
        print(f" {n} is divisible by 5")
    else:
        print(f" {n} is not divisible by 5")
```

```python
check(10)
```

```python
check(-15)
```
