```python
stud = {
    "s1" : {"sub1" : 15,"sub2" : 23 , "sub3" : 17},
    "s2" : {"sub1" : 52,"sub2" : 14 , "sub3" : 43},
    "s3" : {"sub1" : 92,"sub2" : 42 , "sub3" : 54}
}
```

```python
stud
```

```python
type(stud)
```

```python
for ind,val in stud.items():
    for ind1,val1 in val.items():
        print(ind,ind1,val1)
```

```python
# find the avg marks of every stud
for ind,val in stud.items():
    avg_marks = 0
    marks = 0
    n = 0 
    for val1 in val.values():
        n += 1
        marks += val1
        avg_marks = marks / n
    print("avg_marks of" ,ind," : ",avg_marks)
```
