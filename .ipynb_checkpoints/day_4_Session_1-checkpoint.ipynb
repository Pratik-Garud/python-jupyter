```python
def price_per_night(x):
    if x <200:
        print("affordable")
    elif x <400:
        print("mid")
    else:
        print("expensiv")
```

```python
def price_per_night(x):
    if (x <200):
        print("affordable")
    elif (x <400):
        print("mid")
    else:
        print("expensiv")
```
