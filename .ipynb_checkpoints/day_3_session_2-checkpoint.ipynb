# date range

```python
# Q1.i want ot define 10 days date range from 1 oct
```

```python
pd.date_range("2025-10-1",periods =10,freq = "D")        # yyyy-mm-dd
```

```python
# Q2.i want ot define 5 month date range from oct
pd.date_range("2025-10-1",periods = 5 ,freq = "ME")        # yyyy-mm-dd
```

```python
# Q3.i want ot define 5 years date range from FEB
pd.date_range("2025-10-1",periods = 5 ,freq = "YE-FEB")        # yyyy-mm-dd
```
