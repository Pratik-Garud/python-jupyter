### Markdown Key Formulas for Jupyter Notebooks

#### 1. **Headings**
To create headings, use the `#` symbol:
```markdown
# This is an H1 (largest heading)
## This is an H2
### This is an H3
#### This is an H4
##### This is an H5 (smallest heading)
```

#### 2. **Bold and Italic Text**
- **Bold:** Use double asterisks `**` or double underscores `__`:
  ```markdown
  **This is bold text**
  __This is also bold text__
  ```
- *Italic:* Use single asterisk `*` or single underscore `_`:
  ```markdown
  *This is italic text*
  _This is also italic text_
  ```
- **Bold + Italic:** Combine `***`:
  ```markdown
  ***This is bold and italic text***
  ```

#### 3. **Lists**
- **Unordered List:** Use `-`, `*`, or `+` for bullet points:
  ```markdown
  - Item 1
  - Item 2
    - Subitem 2.1
  ```
- **Ordered List:** Use numbers:
  ```markdown
  1. Step 1
  2. Step 2
     1. Substep 2.1
  ```

#### 4. **Links**
Use `[text](URL)` for links:
```markdown
[Visit Google](https://www.google.com)
```

#### 5. **Images**
Use `![alt text](URL or file path)`:
```markdown
![Sample Image](https://via.placeholder.com/150)
```

#### 6. **Code Blocks**
- **Inline Code:** Use backticks `` ` ``:
  ```markdown
  `This is inline code`
  ```
- **Code Block:** Use triple backticks `` ``` `` with the language name (e.g., `python` for syntax highlighting):
  ```python
  def example_function():
      print("Hello, World!")
  ```

#### 7. **Blockquotes**
Use `>` for blockquotes:
```markdown
> This is a blockquote
```

#### 8. **Tables**
Create tables using `|` and `-`:
```markdown
| Column 1 | Column 2 | Column 3 |
|----------|----------|----------|
| Data 1   | Data 2   | Data 3   |
```

#### 9. **Horizontal Line**
Use `---`, `***`, or `___` for a horizontal line:
```markdown
---
```

#### 10. **Mathematical Expressions**
Use `$` for inline math or `$$` for block math (LaTeX syntax):
```markdown
Inline math: $x + y = z$
Block math:
$$
x^2 + y^2 = z^2
$$
```
