### label rotation

```python
df3.plot(x="days",y="values",figsize=(15,5))  #by default it will take color is blue
plot.xlabel("Days of the value")
plot.ylabel("accutal values in millions")
plot.xticks(rotation = 60)         #helps to rotate the lables on x axis
plot.yticks(rotation = 60)         #helps to rotate the lables y axis
```

### color

```python
df3.plot(x="days",y="values",figsize=(15,5),color = "m")  #color is balck
plot.xlabel("Days of the value")
plot.ylabel("accutal values in millions")
plot.xticks(rotation = 60)         #helps to rotate the lables on x axis
plot.yticks(rotation = 60)         #helps to rotate the lables y axis




# m - majenta
# b - blue 
# k - black
# y - yellow
# r - red
# w - white (but you take care of background color)
```

###  line style

```python
df3.plot(x="days",y="values",figsize=(15,5),color = "m",linestyle = "--")  
plot.xlabel("Days of the value")
plot.ylabel("accutal values in millions")
plot.xticks(rotation = 60)        
plot.yticks(rotation = 60)

# by default solid line
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```
